```python
from gensim.corpora import Dictionary

docs = datas["stem_words"].apply(lambda x: x.split())

dictionary = Dictionary(docs)
dictionary.filter_extremes(no_below= 20, no_above= 0.5)

print(dictionary)
```

```python
most_used_word = dictionary.most_common(1)[0]

print(f"Il y a {len(dictionary)} mots uniques")
print(f"Le mot le plus utilisé est {most_used_word[0]} qui apparaît {most_used_word[1]} fois")
```

```python
def keep_tokens(posts, words_list):
    post_words_list = []
    for word in posts:
        if word in words_list:
            post_words_list.append(word)
    
    return " ".join(post_words_list)
```

```python
words_list = set(dictionary.token2id.keys())

datas["imp_words"] = datas["stem_words"].apply(lambda x: keep_tokens(x.split(), words_list))
```

J'effectue un PCA sur le jeu de données pour voir s'il y a des composantes principales significatives.

D'abord, je dois transformer "corpus" en une matrice TFIDF.

```python
samples = np.random.choice(len(datas), 10000, replace= False)
df_samples = datas.iloc[samples, :].reset_index(drop= True)
df_samples
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(token_pattern= r'[\S]+', dtype= np.int8)

bag_of_words = vectorizer.fit_transform(df_samples["imp_words"])

bag_of_words.shape
```

```python
from sklearn.decomposition import TruncatedSVD, PCA

components = 750

tSVD = TruncatedSVD(n_components= components, random_state= 42)
tSVD.fit(bag_of_words)

tSVD_df = pd.DataFrame(tSVD.transform(bag_of_words))
```

```python
tsvd_expl_ratio_sum = tSVD.explained_variance_ratio_.cumsum()

tsvd_expl_ratio_sum[-1:]
```

Pas de composante principale significative pour le PCA.

```python
from sklearn.manifold import TSNE

tsne = TSNE(init= "random", learning_rate= "auto", random_state= 42)
# tsne.fit(bag_of_words)
```

```python
tsne_df = tsne.fit_transform(bag_of_words)
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

k_viz = KElbowVisualizer(KMeans(random_state= 42), k=(3, 25))
k_viz.fit(tSVD_df)
k_viz.poof()
```

```python
kmean = KMeans(n_clusters= k_viz.elbow_value_, random_state= 42).fit(tSVD_df)

tsne_df["labels"] = kmean.labels_
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(tsne_df.loc[:, "PC0"], tsne_df.loc[:, "PC1"], c= tsne_df.loc[:, "labels"], cmap= "hsv")

plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(tSVD_df, tsne_df["labels"])
```

On ne voit pas de clusters significatifs parmis les posts en se basant sur les mots dans les posts même en réduisant les données à des features significatives via le PCA.

Je peux peut-être trouver un meilleur nombre de clusters avec la méthode du silhouette score.

```python
transf_datas_samples = pd.read_csv("P5_06_transf_datas.csv").iloc[samples, :]

transf_datas_samples
```

```python
transf_datas_samples.to_csv("P5_06_samples_datas.csv", index= False)
```

```python
tags_vectorizer = CountVectorizer(token_pattern= r'[\S]+', min_df= 10, dtype= np.int8)

bag_of_tags = tags_vectorizer.fit_transform(transf_datas_samples["Tags"])

tags_df = pd.DataFrame(bag_of_tags.toarray(), columns= tags_vectorizer.get_feature_names_out())

print(tags_df.shape)
tags_df.columns
```

```python
components = 200

tPCA = PCA(n_components= components, random_state= 42)
tPCA.fit(tags_df)

tPCA_df = pd.DataFrame(tPCA.transform(tags_df))
```

```python
tsvd_expl_ratio_sum = tPCA.explained_variance_ratio_.cumsum()

tsvd_expl_ratio_sum[-1:]
```

```python
t_tsne = TSNE(init= "random", learning_rate= "auto", random_state= 42)
```

```python
t_tsne_df = t_tsne.fit_transform(bag_of_tags)
```

```python
t_tsne_df = pd.DataFrame(t_tsne_df, columns= ["PC0", "PC1"])
```

```python
t_k_viz = KElbowVisualizer(KMeans(random_state= 42), k=(3, 25))
t_k_viz.fit(tPCA_df)
t_k_viz.poof()
```

```python
kmean = KMeans(n_clusters= t_k_viz.elbow_value_, random_state= 42).fit(tPCA_df)

t_tsne_df["labels"] = kmean.labels_
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(t_tsne_df.loc[:, "PC0"], t_tsne_df.loc[:, "PC1"], c= t_tsne_df.loc[:, "labels"], cmap= "hsv")

plt.show()
```

On peut voir des clusters pour les tags.

Néanmoins, ils semblent assez dispersés voire séparés.

Je vais mesurer un silhouette score pour me faire une idée de la cohérence des clusters.

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(tPCA_df, t_tsne_df["labels"])

sil_score
```

###### Conclusion

Aucun schéma particulier se dégage des posts.

En ce qui concerne les tags, ils ne sont pas corrélés suivant la corrélation de pearson. Néanmoins, ils apparait qu'ils forment des clusters relativement bien formés même si les individus sont proches des frontiers des clusters (silhouette score == 0.085).
