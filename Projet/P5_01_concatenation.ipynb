Je vais retirer les ponctuations sauf "#", "+" et "." et les stopwords.

* Je retire les "." des fin de phrases et les remplace par un espace si plusieurs à la suite
* Je garde uniquement les "+" de fin de mots et remplace les autre par un espace

Je récupère aussi le nombre de mots pour chaque post.


```python
import re

def ponctuations(post):
    words_list = []
    splited_post = post.split()
    for word in splited_post:
        # word = re.sub(r'-', " ", word)
        word = re.sub(r'\.$', " ", word)
        word = re.sub(r'\.{2,}', " ", word)
        word = re.sub(r'\+{1}', " ", word)
        word = re.sub(r'\'s', " is", word)
        word = re.sub(r'n\'t', " not", word)
        
        words_list.append(word)

    return " ".join(words_list)
```

```python
tmp_df = posts_tagged_df["full_posts"].apply(lambda x: ponctuations(x))

tmp_df
```

```python
from nltk import RegexpTokenizer
from nltk.corpus import stopwords

# full_nb_words = np.zeros(len(posts_tagged_df))
# no_stops_nb_words = np.zeros(len(posts_tagged_df))
# count = 0

def remove_stopwords(post, tokenizer, stops):

    words = tokenizer.tokenize(post)
    words = [w for w in words if not w.isnumeric()]
    words = [w for w in words if len(w) > 1]

    important_words = [x for x in words if x not in stops]
    important_words = [w for w in important_words if not w.isnumeric()]
    important_words = [w for w in important_words if len(w) > 1]

    return np.array([" ".join(words), " ".join(important_words)])
```

```python
stops = set(stopwords.words("english"))

regex_form = r'[a-zA-Z0-9]+\.{1}[a-zA-Z0-9]+|[a-zA-Z]+\#?\+*|[a-zA-Z0-9]+\-*[a-zA-Z0-9]*|\.?[a-zA-Z]+'
tokenizer = RegexpTokenizer(regex_form)

tmp_dff = tmp_df.apply(lambda x: 
    remove_stopwords(x, tokenizer, stops))
```

Je vais maintenant regarder la répartition de "ViewCount" et "AnswerCount" et garder les posts avec le plus d'impact.

```python
viewcount_log = np.log(posts_tagged_df["ViewCount"])
quantile_vals = [0.25, 0.5, 0.75]
q_view_log = np.quantile(viewcount_log, quantile_vals)
q_view = np.quantile(np.round(np.exp(viewcount_log)), quantile_vals)

colors = {0: "r", 1: "y", 2: "r"}

plt.hist(viewcount_log, bins= 25)
for i in range(len(q_view_log)):
    plt.axvline(q_view_log[i], ls= "--", c= colors[i], label= f"Q{quantile_vals[i]}: {q_view[i]}")

plt.title("ViewCount")
plt.legend()

plt.show()

```

Grâce à "TagSynonyms.csv", je transforme les mots des posts similaires à des tags par leur tag associé.

```python
tags_synonyms = pd.read_csv("TagSynonyms.csv")
tags_synonyms
```

```python
counter = 0

def transform_words(splitted_words):
    global counter

    if (counter +1) % 1000 == 0:
        print(f"{counter +1}/{len(posts_tagged_df)}")

    resulting_words = []
    for word in splitted_words:
        if word in tags_synonyms["SourceTagName"].values:
            new_word = str(tags_synonyms.loc[tags_synonyms["SourceTagName"] == word, "TargetTagName"].values[0])
            resulting_words.append(new_word)
        # elif word in tags_synonyms["TargetTagName"].values:
        else:
            resulting_words.append(word)

    counter += 1
    # print(resulting_words)
    return " ".join(resulting_words)
```

```python
posts_tagged_df["no_stops_posts"] = posts_tagged_df["no_stops_posts"].apply(lambda x: transform_words(x.split()))
```
